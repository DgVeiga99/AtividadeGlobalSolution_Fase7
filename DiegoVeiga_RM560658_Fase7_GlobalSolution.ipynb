## ‚Ü™ **DESASTRE - Vale do Itaja√≠ 2008**

**M√™s do ocorrido:** Novembro

**Data do ocorrido:** N/A

### **Leitura dos dados**

```python
# === Leitura dos dados de entrada ===

input_2008 = [
    "INMET_S_SC_A841_JOACABA_01-01-2008_A_31-12-2008.xlsx",
    "INMET_S_SC_A860_CURITIBANOS_28-02-2008_A_31-12-2008.xlsx",
    "INMET_S_SC_A861_RIO DO CAMPO_10-03-2008_A_31-12-2008.xlsx",
    "INMET_S_SC_A863_ITUPORANGA_04-03-2008_A_31-12-2008.xlsx"
    ]

output_2008 = []
```

```python
# === Designa√ß√£o do m√™s de desastre ===

# M√™s com desastre no formato (ano, m√™s)
meses_com_desastre = [(2008, 11)]
meses_interesse = [(2008,10),(2008, 11)]
```

```python
# === Formata√ß√£o dos dados ===

for arquivo in input_2008:

  # L√™ os metadados para obter a cidade
  df_cidade = pd.read_excel(arquivo, nrows=8)
  cidade = df_cidade.iloc[1, 1]

  # L√™ os dados ignorando os metadados
  df_dados = pd.read_excel(arquivo, skiprows=8)

  # Adiciona a cidade em todos os registros
  df_dados["CIDADE"] = cidade

  # Converte a coluna de data para datetime (essencial para comparar ano e m√™s)
  df_dados['DATA (YYYY-MM-DD)'] = pd.to_datetime(df_dados['DATA (YYYY-MM-DD)'])

  # Filtra apenas os dados nas datas de interesse
  df_dados = df_dados[
      df_dados['DATA (YYYY-MM-DD)'].apply(lambda d: (d.year, d.month) in meses_interesse)]

  # Marca como 1 o valor de desastre para datas que est√£o nos meses indicados
  df_dados['DESASTRE'] = df_dados['DATA (YYYY-MM-DD)'].apply(
      lambda d: 1 if (d.year, d.month) in meses_com_desastre else 0
      )

  # Salva o resultado final
  output_2008.append(df_dados)
```
